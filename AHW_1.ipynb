#### 2) Adjusting alpha and marker size to compensate for overplotting

```python
sns.scatterplot(data=df ,x=df.longitude, y=df.latitude , hue= df.daynight , style=df.daynight)
plt.show()
```

```python
sns.scatterplot(data=df ,x=df.longitude, y=df.latitude,hue=df.confidence,palette='plasma')
plt.show()
```

#### 3) Using a hexbin plot.

```python
plt.figure(figsize=(15,10))
plt.hexbin(x=df.longitude, y=df.latitude , bins='log' ,cmap=plt.cm.Greens)
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.colorbar()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
subsample_data =df.sample(frac=0.05)
sns.scatterplot(data=subsample_data ,x=subsample_data.longitude, y=subsample_data.latitude ,color='r' ,alpha=0.20)
plt.title("5% Data")

plt.subplot(2,2,2)
subsample_data =df.sample(frac=0.10)
sns.scatterplot(data=subsample_data ,x=subsample_data.longitude, y=subsample_data.latitude ,color='b' ,alpha=0.30,hue=df.daynight)
plt.title("10% Data")

plt.subplot(2,2,3)
subsample_data =df.sample(frac=0.15)
sns.scatterplot(data=subsample_data ,x=subsample_data.longitude, y=subsample_data.latitude ,color='g',alpha=0.40)
plt.title("15% Data")

plt.subplot(2,2,4)
subsample_data =df.sample(frac=0.20)
sns.scatterplot(data=subsample_data ,x=subsample_data.longitude, y=subsample_data.latitude ,color='k' ,alpha=0.10)
plt.title("20% Data")
```

```python
max_sat = df.bright_ti4.max()
```

```python
max_sat
```

```python
Highsat = df[df['bright_ti4'] >= max_sat]
```

```python
Highsat
```

```python
Lowsat = df[df['bright_ti4'] < max_sat]
```

```python
Lowsat
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
sns.scatterplot(data=Highsat ,y=Highsat['latitude'] , x=Highsat['longitude'] ,color='g')

plt.subplot(2,2,2)
sns.scatterplot(data=Lowsat, y=Lowsat['latitude'] , x=Lowsat['longitude'] ,color='b')
```
